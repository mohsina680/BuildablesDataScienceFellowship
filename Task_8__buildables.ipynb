1. NumPy Array Creation + Basic Operations

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 1D array creation, addition, multiplication, slicing
arr1 = np.array([1, 2, 3, 4, 5])
arr2 = np.array([10, 20, 30, 40, 50])
sum_arr = arr1 + arr2
prod_arr = arr1 * arr2
slice_arr = arr1[1:4]
print("1D sum:", sum_arr, "product:", prod_arr, "slice:", slice_arr)

# 2D array creation and operations
mat1 = np.array([[1, 2], [3, 4]])
mat2 = np.ones((2,2)) * 5
sum_mat = mat1 + mat2
prod_mat = mat1 * mat2
print("2D sum:\n", sum_mat, "\n2D product:\n", prod_mat)
```

2. Array Statistics

```python
mean = np.mean(arr1)
median = np.median(arr1)
std_dev = np.std(arr1)
argmax = np.argmax(arr2)
print("Mean:", mean, "Median:", median, "Std:", std_dev, "Argmax index:", argmax)
```

Visualization with Matplotlib & Seaborn

11. Line Plot Basics

```python
years = [2010, 2011, 2012, 2013]
values = [100, 150, 120, 180]
plt.plot(years, values, marker='o')
plt.xlabel('Year')
plt.ylabel('Value')
plt.title('Yearly Trend')
plt.show()
```

12. Bar Chart Frequencies

```python
categories = ['A', 'B', 'C']
counts = [10, 20, 15]
plt.bar(categories, counts)
plt.xlabel('Category')
plt.ylabel('Frequency')
plt.title('Category Frequencies')
plt.show()
```

13. Scatter Plot Relationships

```python
x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 25, 30]
plt.scatter(x, y)
plt.xlabel('X Variable')
plt.ylabel('Y Variable')
plt.title('Scatter Plot')
plt.show()
```

14. Seaborn Pairplot

```python
tips = sns.load_dataset("tips")
sns.pairplot(tips, hue='sex')
plt.show()
```

15. Category Box Plot

```python
sns.boxplot(x='day', y='total_bill', data=tips)
plt.title('Total bill by day')
plt.show()
```

16. Correlation Heatmap

```python
corr = tips.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

17. Subplot Grid

```python
fig, axes = plt.subplots(1, 2, figsize=(10,4))
axes[0].plot(years, values, marker='o')
axes[0].set_title('Line Plot')
sns.scatterplot(x='total_bill', y='tip', data=tips, ax=axes[1])
axes[1].set_title('Scatter Plot')
plt.tight_layout()
plt.show()
```

18. Seaborn Styling

```python
sns.set_palette('Set2')
sns.set_style('whitegrid')
sns.scatterplot(x='total_bill', y='tip', hue='day', data=tips)
plt.title('Styled Scatter Plot')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()
```

19. Violin Plot Distribution

```python
iris = sns.load_dataset("iris")
sns.violinplot(x='species', y='sepal_length', data=iris)
plt.title('Sepal Length Distribution by Species')
plt.show()
```

20. Hybrid Visualization

```python
data = np.random.randn(1000)
plt.hist(data, bins=30, density=True, alpha=0.6)
sns.kdeplot(data, color='red')
plt.title('Histogram with KDE Overlay')
plt.show()
```
